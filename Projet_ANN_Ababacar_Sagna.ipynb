# Projet 1: ANN

### Les données

Nous utiliserons un sous-ensemble de l'ensemble de données du LendingClub

LendingClub est une société américaine de prêt entre pairs, dont le siège se trouve à San Francisco, en Californie[3]. Elle a été la première société de prêt entre pairs à enregistrer ses offres en tant que titres auprès de la Securities and Exchange Commission (SEC) et à proposer des prêts sur un marché secondaire. LendingClub est la plus grande plateforme de prêt entre pairs au monde.

### Votre objectif

Compte tenu des données historiques sur les prêts accordés avec des informations sur le défaut de paiement (charge-off) ou non de l'emprunteur, pouvons-nous construire un modèle qui puisse prédire si un emprunteur remboursera son prêt ou non ? Ainsi, à l'avenir, lorsque nous aurons un nouveau client potentiel, nous pourrons évaluer s'il est susceptible ou non de rembourser le prêt. Gardez à l'esprit les mesures de classification lorsque vous évaluez la performance de votre modèle !

La colonne "loan_status" contient notre étiquette ou label.

**Tâche : Examinons la corrélation entre les variables de features continues. Calculer la corrélation entre toutes les variables numériques continues en utilisant la méthode .corr().**

```python
# Code ici
# Sélectionner uniquement les colonnes numériques
numeric_columns = df.select_dtypes(include=[np.number]).columns.tolist()

# Calculer la matrice de corrélation
correlation_matrix = df[numeric_columns].corr()

print(f"\nMatrice de corrélation calculée pour {len(numeric_columns)} variables numériques")
print("\nMATRICE DE CORRÉLATION COMPLÈTE :")
print("=" * 200)

# Affichage sans coupure avec to_string()
print(correlation_matrix.to_string(max_cols=None, max_rows=None, line_width=None))
```

**Tâche : Vous devriez avoir remarqué une corrélation presque parfaite avec la feature "installment". Explorez cette feature plus en détail.
```python
# Code ici
# Trouver la variable la plus corrélée avec installment
installment_corr = correlation_matrix['installment'].abs().sort_values(ascending=False)
print("Corrélations avec 'installment' (triées par valeur absolue) :")
print(installment_corr.head(10))
```

```python
# La variable la plus corrélée (après installment lui-même)
most_correlated = installment_corr.index[1]  # [0] est installment lui-même
correlation_value = correlation_matrix.loc['installment', most_correlated]

print(f"\nVariable la plus corrélée avec 'installment' : {most_correlated}")
print(f"Coefficient de corrélation : {correlation_value:.4f}")
```

```python
# Descriptions des variables
print(f"\nDESCRIPTIONS DES VARIABLES :")
print(f"installment :")
print("The monthly payment owed by the borrower if the loan originates.")
print(f"\nloan_amnt :")
print("The listed amount of the loan applied for by the borrower. If at some point in time, the credit department reduces the loan amount, then it will be reflected in this value.")

print(f"\nSTATISTIQUES DESCRIPTIVES :")
print(f"Installment :")
print(df['installment'].describe())
print(f"\n{most_correlated} :")
print(df[most_correlated].describe())
```

```python
# Créer le scatterplot
plt.figure(figsize=(10, 6))
plt.scatter(df['installment'], df[most_correlated], alpha=0.5, s=1)
plt.xlabel('installment', fontsize=12)
plt.ylabel(f'{most_correlated}', fontsize=12)
plt.title(f'Relation entre installment et {most_correlated}\n(r = {correlation_value:.3f})',
         fontsize=12, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Analyse de la relation
print(f"\nANALYSE DE LA RELATION :")
print(f"La corrélation entre 'installment' et '{most_correlated}' est de {correlation_value:.4f}")

if abs(correlation_value) > 0.9:
    print("⚠️  CORRÉLATION TRÈS FORTE (> 0.9) - Possible redondance d'information !")
    print("Cette relation a du sens car :")
    if most_correlated == 'loan_amnt':
        print("- L'acompte mensuel (installment) dépend directement du montant du prêt")
        print("- Plus le montant est élevé, plus l'acompte mensuel est élevé")
        print("- Il s'agit probablement d'informations redondantes")
elif abs(correlation_value) > 0.7:
    print("✅ CORRÉLATION FORTE (0.7-0.9) - Relation logique attendue")
elif abs(correlation_value) > 0.5:
    print("✅ CORRÉLATION MODÉRÉE (0.5-0.7)")
else:
    print("✅ CORRÉLATION FAIBLE (< 0.5)")

print("\n🔍 RECOMMANDATION :")
if abs(correlation_value) > 0.95:
    print(f"Considérez supprimer une des deux variables ({most_correlated} ou installment) pour éviter la multicolinéarité")
else:
    print("Les deux variables peuvent être conservées")
```

**Tâche : Créer un diagramme en barres (bar plot) montrant la corrélation entre les features numériques et la nouvelle colonne loan_repaid. [Lien utile](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.bar.html)**

```python
# Code ici
# Sélectionner les colonnes numériques (exclure loan_repaid pour éviter l'autocorrélation)
numeric_columns = df.select_dtypes(include=[np.number]).columns.tolist()
if 'loan_repaid' in numeric_columns:
    numeric_columns.remove('loan_repaid')

print(f"Variables numériques analysées : {len(numeric_columns)}")
print(numeric_columns)

# Calculer les corrélations avec loan_repaid
correlations = df[numeric_columns + ['loan_repaid']].corr()['loan_repaid'].drop('loan_repaid')

# Trier par valeur absolue pour visualiser les plus importantes
correlations_sorted = correlations.reindex(correlations.abs().sort_values(ascending=False).index)

# Créer le bar plot
plt.figure(figsize=(15, 8))
colors = ['red' if x < 0 else 'green' for x in correlations_sorted.values]
bars = plt.bar(range(len(correlations_sorted)), correlations_sorted.values, color=colors, alpha=0.7)

plt.title('Corrélation entre les features numériques et loan_repaid', fontsize=16, fontweight='bold')
plt.xlabel('Features numériques', fontsize=12)
plt.ylabel('Coefficient de corrélation', fontsize=12)
plt.xticks(range(len(correlations_sorted)), correlations_sorted.index, rotation=45, ha='right')
plt.axhline(y=0, color='black', linestyle='-', alpha=0.3)
plt.grid(True, alpha=0.3, axis='y')

# Ajouter une ligne à y=0 pour référence
plt.tight_layout()
plt.show()
```

**Tâche : On dirait que la feature total_acc est en corrélation avec la feature mort_acc, c'est logique ! Essayons cette approche fillna(). Nous allons regrouper le DataFrame par total_acc et calculer la valeur moyenne de mort_acc par entrée de total_acc. Pour obtenir le résultat ci-dessous:**

```python
# Code ici

# Créer la série des moyennes de mort_acc par total_acc
mean_mort_acc_by_total_acc = df.groupby('total_acc')['mort_acc'].mean()

print("Moyenne de mort_acc par valeur de total_acc :")
print("=" * 45)
print(mean_mort_acc_by_total_acc.head(20))

print(f"\nNombre de valeurs avant remplissage : {df['mort_acc'].count():,}")
print(f"Valeurs manquantes avant : {df['mort_acc'].isnull().sum():,}")

# Fonction pour remplir les valeurs manquantes
def fill_mort_acc(row):
    """
    Remplit les valeurs manquantes de mort_acc basées sur total_acc
    """
    # Check if mort_acc is NaN AND total_acc is NOT NaN
    if pd.isna(row['mort_acc']) and pd.notna(row['total_acc']):
        # Use the mean for this value of total_acc
        # Use .get() with a default (like NaN) in case total_acc value
        # was not in the original groupby index (e.g., if total_acc itself was NaN)
        return mean_mort_acc_by_total_acc.get(row['total_acc'], np.nan)
    else:
        # Keep the existing value (either not NaN or total_acc was also NaN)
        return row['mort_acc']

# Apply the function to fill the missing values
df['mort_acc'] = df[['mort_acc', 'total_acc']].apply(fill_mort_acc, axis=1)

print(f"Nombre de valeurs après remplissage : {df['mort_acc'].count():,}")
print(f"Valeurs manquantes après : {df['mort_acc'].isnull().sum():,}")
print("✅ Valeurs manquantes de mort_acc remplies avec succès !")
```

```python
# Code ici
print("Exemples d'adresses :")
print(df['address'].head(10))

# Fonction pour extraire le code postal (derniers 5 caractères)
def extract_zip_code(address):
    """Extrait le code postal de l'adresse (supposé être les 5 derniers caractères)"""
    if pd.isna(address):
        return np.nan
    # Prendre les 5 derniers caractères après avoir enlevé les espaces
    return str(address).strip()[-5:]

# Créer la colonne zip_code
df['zip_code'] = df['address'].apply(extract_zip_code)

print(f"\nCodes postaux extraits :")
print(f"Nombre de codes postaux uniques : {df['zip_code'].nunique()}")
print(f"Exemples de codes postaux :")
print(df['zip_code'].value_counts().head(10))
```

```python
# Code ici
print(f"Nombre de codes postaux uniques : {df['zip_code'].nunique()}")

# Créer les variables dummy pour zip_code
zip_code_dummies = pd.get_dummies(df['zip_code'], prefix='zip_code', drop_first=True)

print(f"Nombre de variables dummy créées : {len(zip_code_dummies.columns)}")

# Concaténer avec le DataFrame
df = pd.concat([df, zip_code_dummies], axis=1)

# Supprimer les colonnes originales (zip_code et address)
df = df.drop(['zip_code', 'address'], axis=1)

print(f"Taille finale du DataFrame : {df.shape}")
print("✅ zip_code converti en variables dummy, colonnes zip_code et address supprimées !")
```

### Feature earliest_cr_line

**Tâche : Il semble s'agir d'une feature de time stamp historique. Vous pouvez extraire l'année de cette feature à l'aide d'une fonction .apply, puis convertissez-la en une feature numérique. Définissez cette nouvelle donnée dans une colonne feature qu'on appelle 'earliest_cr_year', puis supprimez la feature earliest_cr_line**.

```python
# Code ici
print("Aperçu de earliest_cr_line :")
print(df['earliest_cr_line'].head(10))
print(f"Type de données : {df['earliest_cr_line'].dtype}")

# Fonction pour extraire l'année
def extract_year(date_str):
    """
    Extrait l'année d'une chaîne de date
    Format attendu : 'Jan-1990' ou similaire
    """
    if pd.isna(date_str):
        return np.nan
    try:
        # Diviser par '-' et prendre la dernière partie (année)
        year = int(str(date_str).split('-')[-1])
        return year
    except:
        return np.nan

# Appliquer la fonction pour créer earliest_cr_year
df['earliest_cr_year'] = df['earliest_cr_line'].apply(extract_year)

print(f"\nAperçu de earliest_cr_year :")
print(df['earliest_cr_year'].head(10))
print(f"Type de données : {df['earliest_cr_year'].dtype}")

print(f"\nStatistiques de earliest_cr_year :")
print(df['earliest_cr_year'].describe())

print(f"\nDistribution des années :")
year_counts = df['earliest_cr_year'].value_counts().sort_index()
print(year_counts.head(10))

# Supprimer la colonne originale
df = df.drop('earliest_cr_line', axis=1)

print(f"\nTaille du DataFrame : {df.shape}")
print("✅ earliest_cr_year créé et earliest_cr_line supprimé !")
```

## Normalisation des données

**Tâche : Utiliser une mise à l'échelle MinMaxScaler pour normaliser les données de features X_train et X_test. Rappelez-vous que nous ne voulons pas de fuite de données du dataset, donc nous adaptons que sur les données X_train.**

```python
# Code ici
scaler = MinMaxScaler()
scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Conversion en DataFrame
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns, index=X_train.index)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns, index=X_test.index)

print(f"✅ Normalisation appliquée [0,1]")
print(f"Données normalisées - Min: {X_train_scaled.min().min():.2f}, Max: {X_train_scaled.max().max():.2f}")
```

## Création du Modèle

**Tâche : Exécutez la cellule ci-dessous pour importer les fonctions Keras nécessaires.**

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

**Tâche : Construire un modèle séquentiel qui sera entraîné sur les données. Vous avez ici des options illimitées, mais voici ce que la solution utilise : un modèle schématisé ainsi : 78 --> 39 --> 19--> 1 neurone de sortie. Optionnel : Explorer l'ajout de [couches d'exclusion ou Dropout layers](https://keras.io/layers/core/) [1](https://en.wikipedia.org/wiki/Dropout_(neural_networks)) [2](https://towardsdatascience.com/machine-learning-part-20-dropout-keras-layers-explained-8c9f6dc4c9ab)**

```python
# Code ici
# Architecture du modèle
num_features = X_train_scaled.shape[1]
print(f"Nombre de features d'entrée : {num_features}")

# Construction du modèle séquentiel
model = Sequential()

# Architecture : 78 -> 39 -> 19 -> 1
model.add(Dense(39, activation='relu', input_shape=(num_features,)))
model.add(Dropout(0.2))
model.add(Dense(19, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(1, activation='sigmoid'))

# Compilation
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

print("Architecture du modèle :")
print(f"• Couches : {num_features} → 39 → 19 → 1")
print(f"• Dropout : 20% sur couches cachées")
print(f"• Activation : ReLU (cachées), Sigmoid (sortie)")
print(f"• Optimiseur : Adam")

model.summary()
```

**Tâche : Adapter le modèle aux données d'entraînement pour au moins 25 Epochs. Ajoutez également les données de validation pour le traçage ultérieur. Facultatif : ajouter un batch_size de 256.**

```python
# Code ici

# Entraînement
print("Début de l'entraînement...")
history = model.fit(
    X_train_scaled, y_train,
    epochs=25,
    batch_size=256,
    validation_data=(X_test_scaled, y_test),
    verbose=1
)

print("✅ Entraînement terminé !")

```
